# Introducción

En este trabajo se analizará un dataset de ventas de una tienda retail utilizando la librería NumPy.

El objetivo es conocer mejor la información, calcular ventas totales, promedios y comparar categorías de productos.

## 1. Importar librerías

Primero se importan las librerías que se usarán durante el trabajo.

```python
import numpy as np

print("NumPy cargado correctamente")
```

## 2. Cargar dataset con NumPy

Se carga el archivo CSV utilizando `genfromtxt()`.

```python
datos = np.genfromtxt(
    '/content/retail_sales_dataset.csv',
    delimiter=',',
    skip_header=1,
    dtype=str
)

print("Dataset cargado correctamente")
print("Dimensión del archivo:", datos.shape)
```

## 3. Revisar primeros datos

Se muestran algunos registros para comprobar que cargó bien.

```python
print(datos[:5])
```

## 4. Preprocesamiento básico

Se eliminan espacios vacíos y se preparan columnas importantes.

```python
datos = np.char.strip(datos)

fechas = datos[:,1]
categorias = datos[:,5]
ventas = datos[:,8].astype(float)

print("Datos preparados correctamente")
print("Categorías encontradas:", np.unique(categorias))
```

## 5. Estadísticas generales con NumPy

Se calculan datos básicos de ventas.

```python
print("Estadísticas Generales")
print("Ventas totales:", np.sum(ventas))
print("Promedio ventas:", round(np.mean(ventas), 2))
print("Venta máxima:", np.max(ventas))
print("Venta mínima:", np.min(ventas))
```

## 6. Total de ventas por categoría

Se suman las ventas según categoría de producto.

```python
categorias_unicas = np.unique(categorias)

print("\nTotal ventas por categoría")

for cat in categorias_unicas:
    total = np.sum(ventas[categorias == cat])
    print(cat, ":", total)
```

## 7. Promedio diario de ventas por categoría

Se calcula el promedio diario de ventas para cada categoría.
Primero se buscaron los días en que hubo ventas de cada categoría. Luego se sumaron las ventas por día y finalmente se calculó el promedio.

```python
categorias = np.char.strip(datos[:,5])
fechas = datos[:,1]
ventas = datos[:,8].astype(float)

categorias_unicas = np.unique(categorias)

print("Promedio diario por categoría")

for cat in categorias_unicas:

    dias = np.unique(fechas[categorias == cat])
    totales = []

    for dia in dias:
        total = np.sum(ventas[(categorias == cat) & (fechas == dia)])
        totales.append(total)

    print(cat, ":", round(np.mean(totales), 2))
```

## 8. Categoría con mayor y menor venta

```python
totales = []

for cat in categorias_unicas:
    total = np.sum(ventas[categorias == cat])
    totales.append(total)

totales = np.array(totales)

mayor = categorias_unicas[np.argmax(totales)]
menor = categorias_unicas[np.argmin(totales)]

print("Categoría con mayores ventas:", mayor)
print("Categoría con menores ventas:", menor)
```

## 9. Filtrar una categoría específica

Se limpiaron los textos de la columna categoría para eliminar espacios y caracteres ocultos.

Luego, se filtraron los registros correspondientes a la categoría Beauty.

```python
categorias = np.char.strip(datos[:,5])
categorias = np.char.replace(categorias, '\r', '')
categorias = np.char.replace(categorias, '\n', '')
```

```python
filtro = datos[categorias == "Beauty"]

print("Primeros registros categoría Beauty:")

print(filtro[:5])
```

## 10. Operaciones básicas

Se realizan operaciones matemáticas solicitadas.

```python
print("Suma:", np.sum(ventas))
print("Resta ejemplo:", ventas[0] - ventas[1])
print("Multiplicación ejemplo:", ventas[0] * 2)
print("División ejemplo:", round(ventas[0] / ventas[1], 2))
```
