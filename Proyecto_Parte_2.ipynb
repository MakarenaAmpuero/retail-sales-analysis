# Introducción

En este proyecto se trabajó con un dataset de ventas de una tienda retail con el objetivo de analizar la información y entender cómo se comportan los datos.

Para esto se utilizaron herramientas como NumPy y Pandas, que permiten manipular y explorar datos de forma más ordenada. A lo largo del trabajo se fueron aplicando distintas operaciones y filtros, lo que ayudó a ir entendiendo mejor el dataset y practicar lo visto en clases.

## 1. Importar librería

Primero se importa Pandas, librería especializada en análisis de datos.

```python
import pandas as pd

print("Pandas cargado correctamente")
```

## 2. Cargar dataset

Se carga el archivo CSV en un DataFrame, ya que Pandas facilita el análisis y manipulación de datos tabulares.

```python
df = pd.read_csv('/content/retail_sales_dataset.csv')

print("Dataset cargado correctamente")
print("Dimensión del archivo:", df.shape)
print("Primeras 10 filas del dataset:")

display(df.head(10))
```

## 3. Exploración inicial de los datos

Se revisan las primeras y últimas filas del dataset para comprobar que la información fue cargada correctamente y observar la estructura general de los registros.

Además, se utiliza `info()` para conocer tipos de datos y detectar posibles valores nulos, mientras que `describe()` permite resumir estadísticamente las columnas numéricas.

```python
#Se muestran las primeras filas para verificar que las columnas y valores fueron importados correctamente.
print("Primeras 10 filas:")
display(df.head(10))
```

```python
#Se muestran las últimas filas para revisar el cierre del dataset y detectar posibles inconsistencias.
print("Últimas 5 filas:")
display(df.tail(5))
```

```python
#Se utiliza para conocer cantidad de registros, tipos de datos y presencia de valores nulos.
print("Información general:")
df.info()
```

```python
#Permite obtener medidas estadísticas como promedio, mínimo, máximo y percentiles.
print("Estadísticas descriptivas:")
df.describe()
```

## 4. Inspección de los datos

Se revisan tipos de datos, valores únicos y frecuencia de categorías.

```python
df.dtypes
```

```python
#Se usa para conocer la frecuencia de cada categoría de producto.
df['Product Category'].value_counts()
```

```python
#Permite identificar los valores únicos presentes en una columna.
df['Customer ID'].unique()
```

## 5. Filtrado de datos

Se aplican filtros para analizar subconjuntos específicos de información según condiciones relevantes.

```python
df[df['Total Amount'] > 50]
```

```python
df[df['Price per Unit'] < 0.5]
```

```python
df.query("`Product Category` == 'Beauty' and `Total Amount` > 30")
```

## 6. Slicing de datos

Se seleccionan filas y columnas específicas del DataFrame.

```python
df[['Product Category', 'Total Amount']]
```

```python
#Se utiliza loc[] para seleccionar filas y columnas por nombre.
df.loc[5:10, ['Product Category', 'Customer ID']]
```

```python
#Se utiliza iloc[] para seleccionar posiciones numéricas dentro del DataFrame.
df.iloc[:5, :3]
```
