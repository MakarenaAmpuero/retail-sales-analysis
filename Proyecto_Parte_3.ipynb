# Introducción

En esta tercera parte del proyecto se continúa trabajando con el dataset de ventas retail, aplicando herramientas más avanzadas de Pandas.

El objetivo es transformar los datos, crear nuevas variables y realizar análisis más completos mediante agrupaciones y funciones personalizadas. Esto permite obtener información más detallada y preparar mejor los datos para futuros análisis.

# 1. Importar librerías

Primero se importa Pandas, librería especializada en análisis de datos.

```python
import pandas as pd

print("Pandas cargado correctamente")
```

# 2. Cargar Dataset

Se carga el archivo CSV en un DataFrame, ya que Pandas facilita el análisis y manipulación de datos tabulares.

```python
df = pd.read_csv('/content/retail_sales_dataset.csv')

print("Dataset cargado correctamente")

print("Dimensión del archivo:", df.shape)

print("Primeras 10 filas del dataset:")
display(df.head(10))

print("Información general del dataset:", df.info)

print("Estadísticas descritivas:", df.describe)

print("Últimas 5 filas del dataset:")
display(df.tail(5))
```

# 3. Transformación de datos
Se crean nuevas columnas para facilitar el análisis y obtener información adicional a partir de los datos existentes.

```python
# Nueva columna de ingreso total
df['Ingreso Calculado'] = df['Quantity'] * df['Price per Unit']
```

```python
# Se normalizan las ventas para poder compararlas en una misma escala.
df['Ventas Normalizadas'] = (
    (df['Total Amount'] - df['Total Amount'].min()) /
    (df['Total Amount'].max() - df['Total Amount'].min())
)
```

```python
# Se clasifican las ventas en categorías para facilitar su interpretación.
def clasificar_ventas(x):
    if x > 1000:
        return 'Alta'
    elif x > 500:
        return 'Media'
    else:
        return 'Baja'

df['Clasificación Venta'] = df['Total Amount'].apply(clasificar_ventas)
```

# 4. Agrupación y agregación
Se agrupan los datos para analizar el comportamiento de las ventas según distintas categorías.

```python
# Agrupación por categoría de producto
df.groupby('Product Category')['Total Amount'].agg(['sum', 'mean', 'count', 'min', 'max', 'std', 'var'])
```

```python
# Agrupación por categoría y género
df.groupby(['Product Category', 'Gender'])['Total Amount'].sum()
```

```python
# Agrupación por mes
df['Date'] = pd.to_datetime(df['Date'])
df['Mes'] = df['Date'].dt.month

df.groupby('Mes')['Total Amount'].sum()
```

# 5. Análisis con apply
Se utilizan funciones personalizadas para obtener información más específica.

```python
# Desviación respecto al promedio del grupo
media_categoria = df.groupby('Product Category')['Total Amount'].transform('mean')

df['Desviación'] = df['Total Amount'] - media_categoria

display(df[['Product Category', 'Total Amount', 'Desviación']].head())
```

El uso de funciones como groupby y apply permitió obtener información más detallada sobre las ventas, lo que ayuda a entender mejor el comportamiento de los datos. Además, estas herramientas son clave para futuros análisis más avanzados en ciencia de datos.
